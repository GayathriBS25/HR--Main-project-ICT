# EXPLORATORY DATA Analysis

Univariate Analysis

```python
hr = df.copy()
sns.set(rc = {'figure.figsize' : (12,6)})
sns.lineplot(x= "Age", y = "Salary", data = hr)
plt.show()
```

We can see that, age ranging from 40 to 70 shows no much variation in the salary, whereas the age gap between 15 and 33 shows greater variations in the salary

```python

sns.set(rc = {'figure.figsize' : (12,6)})
sns.lineplot(x= "Education", y = "Salary", data = hr)
plt.show()
```

Here we observe that, there is drastic increase in salary when education is between Intermediate and Graduation, and laest change in salary when the education is below intermediate.

```python

sns.set(rc = {'figure.figsize' : (12,6)})
sns.lineplot(x= "Salary", y = "Bonus", data = hr)
plt.show()
```

Salary directly propotioanl to Bonus

### *Bivariate Analysis*

```python
plt.figure(figsize=(18,10))
sns.scatterplot(x = "Months", y = "Salary",data = hr)
```

Theres is a commentable hike in salary for newly joined and for the experienced employee


Also we can observe that, there is only a slight change in salary for the employees who stay.

```python
plt.figure(figsize=(18,10))
sns.scatterplot(x = "Bonus", y = "Unit_Sales",data = hr)
```

For the Unit_Sales of <=20, the bonus is set to <4000

For the Unit_Sales of above 20, theres is greater impact in bonus.
